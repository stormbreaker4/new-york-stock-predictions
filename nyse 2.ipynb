```python
u = np.mean(x_train, axis = 0)
std = np.std(x_train, axis = 0)
```

```python
x_train = (x_train-u)/std
x_test = (x_test-u)/std
```

```python
x_train.insert(loc=0, column='ones', value=1)
x_test.insert(loc=0, column='ones', value=1)
```

```python
def hypothesis(x, theta):
    y_ = 0
    n = x.shape[0]

    for i in range(n):
        y_ += theta[i]*x[i]

    return y_
```

```python
def cost(X, Y, theta):
  m = x_train.shape[0]
  total_error = 0

  for i in range(m):
    y_p = hypothesis(X[i], theta)

    total_error += (y_p - Y[i])**2

  return total_error/m
```

```python
def gradient(X, Y, theta):
  m, n = X.shape

  grad = np.zeros((n,))

  for j in range(n):
    for i in range(m):

      x = X[i]

      y_p = hypothesis(x, theta)

      y_t = Y[i]

      grad[j] += (y_p - y_t)*x[j]
    
  return grad/m
```

```python
def gradient_descent(X, Y, learning_rate = 0.001, max_steps = 3000):

  m, n = x_train.shape
  theta = np.zeros((n, ))
  cost_epoch = []

  for i in range(max_steps):
    print("ON: ", i)
    grad = gradient(X, Y, theta)

    e = cost(X, Y, theta)
    cost_epoch.append(e)

    for j in range(n):

      theta[j] = theta[j] - learning_rate*grad[j]

  return theta, cost_epoch
```

```python
def R2_score(y_t, y_p):

  numerator = np.sum((y_t - y_p)**2)
  denominator = np.sum((y_t - y_t.mean())**2)

  score = (1- numerator/denominator)

  return score
```
